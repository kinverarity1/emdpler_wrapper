```python
from eminduction import emdpler
```

```python
r = 4.49  # Coil spacing 
h = 1.0   # Height of coils above the ground
res = [1, 10]      # Resistivity of layered-earth model
thk = [5, 20]           # Thickness of layers (the last is a halfspace)
f1 = 1e6    # Highest frequency
f0 = 1e2    # Lowest frequency
rf, ri = emdpler.vmd(r, src_height=h, rec_level=h, res=res, thk=thk, freq_h=f1, freq_l=f0, 
                     print_input_files=True, calc_displ_currs=True)
```

```python
%matplotlib inline

emdpler.plot_results(rf, ri)
```

```python
print "rf", type(rf)       # the eminduction.emdpler.results class is inherited from a dictionary
for key, value in rf.items():
    print "\t", key.ljust(10), type(value), "shape =", value.shape, 
    print ("min = %.2g" % min(value)).ljust(15), ("max = %.2g" % max(value)).ljust(15)
print "ri", type(ri)
for key, value in ri.items():
    print "\t", key.ljust(10), type(value), "shape =", value.shape, 
    print ("min = %.2g" % min(value)).ljust(15), ("max = %.2g" % max(value)).ljust(15)
```
