```python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
%matplotlib inline
import numpy

from emdpler_wrapper import emdpler, mu0
```

```python
fig = plt.figure(figsize=(14, 3))
gs = gridspec.GridSpec(1, 2, wspace=0.4)
axamp = fig.add_subplot(gs[0])
axph = fig.add_subplot(gs[1])

coil_spacing = 4.49
h = 1.0
r2 = 1
for r1, fmt in [(1, dict(lw=0.7, color="k")),
               (10, dict(lw=1.3, color="k")),
               (100, dict(lw=2, color="k"))]:
    rf, ri = emdpler.vmd(coil_spacing, h, h, res=[r1, r2], thk=[2], freq_h=1e6, freq_l=1e2, calc_displ_currs=True)
    axamp.plot(rf.freq, rf.ampl * mu0 * 1e9, label=r"$\rho_1= %s$ $\Omega$m" % r1, **fmt)
    axph.plot(rf.freq, rf.phase, label=r"$\rho_1= %s$ $\Omega$m" % r1, **fmt)

for ax in (axamp, axph):
    ax.set_xscale("log")
    ax.yaxis.get_major_formatter().set_powerlimits((-2, 3))
    ax.legend(loc="best", fontsize="large", frameon=True)
    ax.set_xlabel("Frequency [Hz]")
    ax.text(0.08, 0.5, r"$\rho_2=%s$ $\Omega$m" % r2, transform=ax.transAxes, ha="left", va="top", fontsize="large")
axamp.set_ylabel(r"Bz amplitude [$\mu$T]")
axph.set_ylabel("Bz phase")
```
